## Data Validation ##


```python
def validate_dataset(df):
    df_temp = df.copy()
    # Check for values < 0 in each numerical column
    print("Checking numeric columns for negative values (< 0):")
    num_cols = df.select_dtypes(include=['int64', 'float64']).columns
    has_negative = False

    for col in num_cols:
        invalid_rows = df_temp[df_temp[col] < 0]

        total_invalid = len(invalid_rows)
        if total_invalid > 0:
            has_negative = True
            print(f"Column '{col}' has {total_invalid} negative values.")

    if not has_negative:
        print("No negative values found in numeric columns.")

    # Blank checking
    print("\nChecking string columns for whitespace-only cells:")
    str_cols = df.select_dtypes(include='object').columns

    if len(str_cols) > 0:
        blankspace_ratio = {
            col: df[col].dropna().apply(lambda x: isinstance(x, str) and x.strip() == '').mean()
            for col in str_cols
        }
        # Print results
        for col, ratio in sorted(blankspace_ratio.items(), key=lambda x: x[1], reverse=True):
            print(f"{col}: {ratio:.2%} of cells contain only whitespace")
    else:
        print(" No string columns found to check whitespace.")

    return df_temp
```

## Handle missing data ##


```python
def handle_missing_values(df):

    df_temp = df.copy()
    initial_shape = df_temp.shape

    # DROP HIGH-NULL OR POST-DELIVERY LEAKAGE COLUMNS
    # Ensure 'REASON_CD' is dropped along with 'SOUF_RCV_NO' and 'QTUF_RCV_NO'
    cols_to_drop = ['SOUF_RCV_NO', 'QTUF_RCV_NO', 'REASON_CD']
    existing_drop_cols = [c for c in cols_to_drop if c in df_temp.columns]

    if existing_drop_cols:
        df_temp = df_temp.drop(columns=existing_drop_cols)
        print(f"Dropped columns: {existing_drop_cols}")
    else:
        print("Target columns to drop are already removed or do not exist.")

    # IMPUTE 'Ship Mode' AND 'SHIP DECISION NO' WITH MODE ---
    # Impute columns that have low-null ratios (2-3%) using the most frequent value
    impute_cols = ['Ship Mode', 'SHIP DECISION NO']
    existing_impute_cols = [c for c in impute_cols if c in df_temp.columns]

    if existing_impute_cols:
        print(f"\nImputing missing values with MODE for: {existing_impute_cols}")

        # Display the number of nulls before imputation
        for col in existing_impute_cols:
            null_count = df_temp[col].isnull().sum()
            if null_count > 0:
                print(f"'{col}' has {null_count} nulls ({null_count/len(df_temp):.2%})")

        # Initialize SimpleImputer with 'most_frequent' strategy
        mode_imputer = SimpleImputer(strategy='most_frequent')

        # Perform the imputation
        df_temp[existing_impute_cols] = mode_imputer.fit_transform(df_temp[existing_impute_cols])
        print("Imputation completed successfully.")
    else:
        print("\nNo target columns found for mode imputation.")

    # --- DATASET SHAPE SUMMARY & RESIDUAL NULL CHECK ---
    print(f"\nSummary of shape change:")
    print(f"Original Shape: {initial_shape}")
    print(f"Current Shape : {df_temp.shape}")

    # Check if there are any remaining null values left in other columns
    remaining_nulls = df_temp.isnull().sum()
    remaining_nulls = remaining_nulls[remaining_nulls > 0]

    if len(remaining_nulls) > 0:
        print(f"\nWarning: There are still missing values left in the dataset:")
        print(remaining_nulls)
    else:
        print("\nNo missing values left in the dataset.")

    return df_temp
```
